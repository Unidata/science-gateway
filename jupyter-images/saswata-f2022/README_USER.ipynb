# JupyterHub Server on the Unidata Science Gateway

### The Terminal

JupyterLabs come equipped with a fully fledged Linux Terminal (`File → New → Terminal`). If there is a package/program you would like to see installed please see the `Issues and Support` section below. 

### Creating/Updating Kernels Per User

For details on the pre-configured environment for this JupyterHub server open the file located at `/README.md` using a text editor (`vim` and `nano` are installed in this image) or file viewer (`cat`, `more`, `less`, etc).

The terminal can be used to create new or update existing iPython environments. Note, however, that any changes are local to this user only. If multiple people use this JupyterHub and would all like an updated environment, contact us with a list of new dependencies (see the `Issues and Support` section).

Our recommended method of installing new dependencies to your environment is by creating an `environment.yml` and then running an update command using either `conda` or `mamba`. The environment files used to create the pre-configured environment are found in `/environment` and can be used as a template. The packages `nb_conda_kernels` and `ipykernel` are necessary to make the environment available to JupyterLab.

```shell
# Copy environment file
cp /environment/environment.yml $HOME

# Edit
<vim/nano> $HOME/environment.yml

# Example environment.yml:
name: <env-name>
channels:
  - conda-forge
dependencies:
  - nb_conda_kernels
  - ipykernel
  - <new-package-1>
  - <new-package-2>
  - <new-package-3>

# Run the update command on your environment using your edited .yml file as input
# NOTE: mamba is a reimplementation of the conda package manager that is usually faster
# NOTE: and often provides more meaningful output. Use whichever you like.
<conda/mamba> env update --name <env-name> -f $HOME/environment.yml
```

If a package is not found in the `conda` repositories, you may have to install this package through `pip`. To do so, edit your `environment.yml` to look like:

```shell
# Example environment.yml with additional pip dependencies:
name: <env-name>
channels:
  - conda-forge
dependencies:
  - nb_conda_kernels
  - ipykernel
  - <new-package-1>
  - <new-package-2>
  - <new-package-3>
  - pip:
    - <pip-package-1>
    - <pip-package-2>
    - <pip-package-3>
```

Then run the same update command found above.

Once the update command has executed successfully, you can make the kernel available by restarting the server (File --> Hub Control Panel --> Stop My Server --> Start My Server) and confirm that the new environment is available in the Launcher (File --> New Launcher).
